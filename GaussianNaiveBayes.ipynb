```python
gnb = GaussianNB()
```

```python
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_samples = 200, n_features =2, n_informative=2, n_redundant=0)
```

```python
plt.scatter(X[:,0], X[:,1],c=Y)
```

```python
print(X[0])
print(X.shape)
```
